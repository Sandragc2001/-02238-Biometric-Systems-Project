```python
import os
from PIL import Image, ImageEnhance

def brighten_images(source_dir, target_dir, brightness_factor):
    
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    #Loop through all files 
    for filename in os.listdir(source_dir):
        if filename.endswith(('.jpg', '.jpeg', '.png')):  #Check for image files
            
            full_path = os.path.join(source_dir, filename)
            
            image = Image.open(full_path)
            #Enhance the brightness
            enhancer = ImageEnhance.Brightness(image)
            img_brightened = enhancer.enhance(brightness_factor)
            #Full path for saving
            save_path = os.path.join(target_dir, filename)
            #Save image
            img_brightened.save(save_path)
            print(f'Brightened image saved to {save_path}')


source_directory = r'C:\Users\sandr\Documents\DINAMARCA\MASTER\Summer2024\trabajo\Final\bonafide_probe'
target_directory = r'C:\Users\sandr\Documents\DINAMARCA\MASTER\Summer2024\trabajo\Final\bonafide_probe_light'
brightness_factor = 1.5  #Increase the brightness by 50%

brighten_images(source_directory, target_directory, brightness_factor)
```
